```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
#Load data
df = pd.read_csv("insurance.csv")  # or pd.read_csv("your_file.csv")
df.head()
```

```python
#overview
df[['age', 'bmi', 'charges']].describe()
```

```python
#Scatter Plot: Age and BMI
sns.scatterplot(x='age', y='bmi', data=df)
plt.title("Age and BMI")
plt.show()
```

```python
#Scatter Plot: Age and Charges
sns.scatterplot(x='age', y='charges', data=df)
plt.title("Age and Charges")
plt.show()
```

```python
#Linear Regression: Age and BMI
X1 = df[['age']]
y1 = df['bmi']

model1 = LinearRegression()
model1.fit(X1, y1)

print("--- Age and BMI ---")
print(f"Coefficient: {model1.coef_[0]}")
print(f"Intercept: {model1.intercept_}")
print(f"R Score: {model1.score(X1, y1)}")
```

```python
#Linear Regression: Age and Charges
X2 = df[['age']]
y2 = df['charges']

model2 = LinearRegression()
model2.fit(X2, y2)

print("--- Age and Charges ---")
print(f"Coefficient: {model2.coef_[0]}")
print(f"Intercept: {model2.intercept_}")
print(f"R Score: {model2.score(X2, y2)}")
```
